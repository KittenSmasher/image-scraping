```python
import pandas as pd

df = pd.read_csv('Music Info.csv')
df.head()
```

```python
with open('track_id.txt', 'r') as file:
    track_ids = [line.strip() for line in file]
    
filtered = df[df['track_id'].isin(track_ids)]
filtered.head()
```

```python
selected = filtered[['track_id', 'spotify_id', 'genre']]
selected.head()
```

```python
selected = selected.sort_values(by='genre')
selected.head()
```

```python
selected.value_counts
```

```python
selected.to_csv('filtered_info.csv', index=False)
```
